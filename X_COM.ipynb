## Import Libraries

```python
import tweepy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

## API Setup

## Fetch Tweets with the Hashtags

```python
# --- STEP 1: Set up Twitter API v2 authentication ---
bearer_token = "YOUR_BEARER_TOKEN"
client = tweepy.Client(bearer_token=bearer_token)
```

```python
# --- STEP 2: Search recent tweets using Twitter API v2 ---
query = "#ALX_AiSK -is:retweet lang:en"
MAX_TWEETS = 10
tweets = client.search_recent_tweets(query=query, max_results=MAX_TWEETS, tweet_fields=["created_at", "public_metrics", "text", "author_id"])
```

```python
# --- STEP 3: Extract relevant tweet data ---
tweet_data = []
for tweet in tweets.data:
    tweet_data.append({
        "created_at": tweet.created_at,
        "text": tweet.text,
        "likes": tweet.public_metrics["like_count"],
        "retweets": tweet.public_metrics["retweet_count"],
        "author_id": tweet.author_id
    })

df = pd.DataFrame(tweet_data)
```

```python
# --- STEP 4: Time Series Analysis ---
df['created_at'] = pd.to_datetime(df['created_at'])
df.set_index('created_at', inplace=True)
df_resampled = df['text'].resample('H').count()

plt.figure(figsize=(10, 5))
df_resampled.plot()
plt.title("Tweet Frequency Over Time (#ALX_AiSK)")
plt.ylabel("Number of Tweets")
plt.xlabel("Time (Hourly)")
plt.grid(True)
plt.show()
```

```python
# --- STEP 5: Word Cloud ---
text_combined = " ".join(df['text'].tolist())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_combined)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud for #ALX_AiSK")
plt.show()
```
